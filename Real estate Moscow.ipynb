```python
import folium
```

```python
data = pd.read_csv('data_set.csv', sep = ',')
moscow_map = folium.Map(location=[55.7558, 37.6173], zoom_start=10)
for index, row in data.iterrows():
    folium.Marker(
        location=[row['geo_lat'], row['geo_lon']],
        popup=f"Property Info: {row['rooms']}",  # Используйте соответствующую колонку для информации о недвижимости
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(moscow_map)
```
